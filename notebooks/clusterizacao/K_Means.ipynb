# 0.0 Imports

```python
import warnings

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn import metrics as mt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
warnings.filterwarnings('ignore')
```

# 1.0 Load Datasets

```python
data_path = '../clusterizacao/X_dataset.csv'

X = pd.read_csv( data_path )
```

# 2.0 Performance

### Parameters
- `K`

## 2.1 Elbow Method

```python
n_clusters = np.arange(2, 11)
init = "k-means++"
n_init = 10

ss_list = []
inertia_list = []
```

```python
# Find best K
for c in n_clusters:
    kmeans = KMeans(
        n_clusters=c,
        init=init,
        n_init=n_init,
        random_state=42,
    )

    # Prevendo
    labels = kmeans.fit_predict(X)

    # Testando performance
    ss = mt.silhouette_score(X, labels)
    ss_list.append(ss)

    inertia_list.append(kmeans.inertia_)

    print(f"For n_clusters = {c} - The Average SS = {ss}")
```

```python
plt.plot(n_clusters, inertia_list, marker="o", label="Inertia")
plt.xlabel("N Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
plt.plot(n_clusters, ss_list, marker="o", label="Inertia")
plt.xlabel("N Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

## 2.2.1 Final Model

```python
best_k = 3

# Treinando e Prevendo
kmeans = KMeans(n_clusters=best_k, init="k-means++", n_init=10)
labels = kmeans.fit_predict(X)

# Testando performance
final_ss = mt.silhouette_score(X, labels)
final_ss
```

## 2.2.2 PCA

```python
# Reduce X
pca = PCA(n_components=2)
reduced_X = pca.fit_transform(X)
reduced_X = pd.DataFrame(data=reduced_X, columns=["PC1", "PC2"])
reduced_X.head()

# reduced centers
centers = pca.transform(kmeans.cluster_centers_)
centers

# Scatter Plot
plt.scatter(reduced_X["PC1"], reduced_X["PC2"], c=kmeans.labels_)
plt.scatter(centers[:, 0], centers[:, 1], marker="x", s=100, c="red")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("K-Means")
```

# 3.0 Results

```python
train_metrics = {
    "Algorithm": "K-Means",
    "N Clusters": best_k,
    "Average Silhouette Score": np.round(final_ss, 3),
}

pd.DataFrame(train_metrics, index=[0]).to_csv("./clustering_metrics.csv")
```
